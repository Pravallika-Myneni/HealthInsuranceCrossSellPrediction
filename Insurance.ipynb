## Analyzing each column

### **Age:**

```
import plotly.express as px
fig = px.histogram(train_df, x="Age")
fig.show()
```

### **Gender:**

```
print("Percentage of males",round(100 * len(train_df[train_df['Gender']=='Male']) / len(train_df), 2))
print("Percentage of females",round(100 * len(train_df[train_df['Gender']=='Female']) / len(train_df), 2))
```

### **Driving_License**

```
list(set(train_df['Driving_License']))
```

```
print("Yes: ",len(train_df[train_df['Driving_License']==1]))
print("No: ",len(train_df[train_df['Driving_License']==0]))
```

###**Region Code**

```
list(set(train_df['Region_Code']))
```

```
fig = px.histogram(train_df, x="Region_Code",nbins=53)
fig.show()
```

### **Previously_Insured :**

```
print("Percentage of Yes:",round(100 * len(train_df[train_df['Previously_Insured']==1]) / len(train_df), 2))
print("Percentage of No: ",round(100 * len(train_df[train_df['Previously_Insured']==0]) / len(train_df), 2))
```
